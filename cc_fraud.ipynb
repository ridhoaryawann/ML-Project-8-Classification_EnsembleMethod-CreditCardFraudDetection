### **Data Preparation**

```python
# Feature Target Split
X = df.drop(['id','Class'], axis=1, errors='coerce')
y = df['Class']

X.head()
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

#### Make Pipeline for Data Transformation

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

col = ['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10', 'V11',
       'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20', 'V21',
       'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']

prep = ColumnTransformer(
    transformers=[
    ('num', scaler, col)
     ],
     remainder = 'passthrough'
)

prep 
```

### **Model Building**

##### DT model

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(
    max_depth=5,
    min_samples_split=10,
    random_state=42
)
```

```python
dt_model = Pipeline(
    [
        ('transformer',prep),
        ('dt',dt)
    ]
)

dt_model.fit(X_train,y_train)
```

##### RF Model

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    min_samples_split=5,
    random_state=42
)
```

```python
rf_model = Pipeline(
    [
        ('transformer', prep),
        ('rf', rf)
    ]
)

rf_model.fit(X_train, y_train)
```

### **Evaluation**

#### 1-Evaluation with cross validation
    splitting data into separate piece to show how the model perform

```python
from sklearn.model_selection import cross_val_score 

# dt 
cv_dt = cross_val_score(dt_model, X_train, y_train, cv=5, scoring='f1')
print(f"Avg CV score for decision tree {np.mean(cv_dt)} | {cv_dt}")
```

```python
# rf
cv_rf = cross_val_score(rf_model, X_train, y_train, cv=5, scoring='f1')
print(f"Avg CV score for random forest : {np.mean(cv_rf)} | {cv_rf}")
```

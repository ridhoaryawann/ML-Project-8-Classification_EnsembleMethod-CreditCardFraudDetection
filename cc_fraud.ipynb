### **Data Preparation**

```python
# Feature Target Split
X = df.drop(['id','Class'], axis=1, errors='coerce')
y = df['Class']

X.head()
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

#### Make Pipeline for Data Transformation

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

col = ['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10', 'V11',
       'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20', 'V21',
       'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']

prep = ColumnTransformer(
    transformers=[
    ('num', scaler, col)
     ],
     remainder = 'passthrough'
)

prep 
```

### **Model Building**

##### DT model

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(
    max_depth=5,
    min_samples_split=10,
    random_state=42
)
```

```python
dt_model = Pipeline(
    [
        ('transformer',prep),
        ('dt',dt)
    ]
)

dt_model.fit(X_train,y_train)
```

##### RF Model

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    min_samples_split=5,
    random_state=42
)
```

```python
rf_model = Pipeline(
    [
        ('transformer', prep),
        ('rf', rf)
    ]
)

rf_model.fit(X_train, y_train)
```

### **Evaluation**

#### 1-Evaluation with cross validation
    splitting data into separate piece to show how the model perform

```python
from sklearn.model_selection import cross_val_score 

# dt 
cv_dt = cross_val_score(dt_model, X_train, y_train, cv=5, scoring='f1')
print(f"Avg CV score for decision tree {np.mean(cv_dt)} | {cv_dt}")
```

```python
# rf
cv_rf = cross_val_score(rf_model, X_train, y_train, cv=5, scoring='f1')
print(f"Avg CV score for random forest : {np.mean(cv_rf)} | {cv_rf}")
```

#### 2-Evaluation with target
    Predict using model and evaluate base on the result

```python
y_pred_rf = rf_model.predict(X_test)

y_pred_dt = dt_model.predict(X_test)
```

##### 2.A. Classification Report
    shows almost all of the basic classification metrics (precision,recall,f1 score)

```python
# rf clss report
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_rf))
```

```python
print(classification_report(y_test, y_pred_dt))
```

##### 2.B. Confusion Matrix
    visualizing classification result of pred and real using Thrue-False-Pos-Neg Matrix

```python
from sklearn.metrics import confusion_matrix

cm_rf = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues')
plt.ylabel('True label')
plt.xlabel('Predicted label')
```

```python
cm_dt = confusion_matrix(y_test, y_pred_dt)
sns.heatmap(cm_dt, annot=True, fmt='d')
plt.ylabel('Actual')
plt.xlabel('Predicted')
```

##### 2.C. Feature Importance 
    See the importance features and evaluate each feature performance to predict

```python
imp_rf = rf_model.named_steps['rf'].feature_importances_
imp_rf_df = pd.DataFrame({
    'Feature' : X.columns, 
    'Importance' : imp_rf
    }).sort_values('Importance',ascending=False)

imp_rf_df.head(5).plot(kind='barh')
```

##### 2.E. ROC / AUC

**Show how model perform under many treshold** not only 0.5 like other matrix

1. ROC (Receiver Operating Characteristic) Curve shows how the true positive rate (recall) vs. false positive rate changes at every possible threshold.

2. AUC (Area Under the Curve) is a single number summarizing the modelâ€™s ability to separate classes, no matter what threshold you choose.

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import auc

y_pred_prob         = rf_model.predict_proba(X_test)[:,1]
fpr, tpr, tresholds = roc_curve(y_test,y_pred_prob)
roc_auc             = auc(fpr, tpr)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label='ROC Curve')
plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')  # diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.4f})', color='blue')
plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve with AUC')
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
plt.show()
```
